# Проект "Прогнозирование оттока клиентов"

Оператор связи «Ниединогоразрыва.ком» хочет научиться прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.

Оператор предоставляет два основных типа услуг: 

- Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
- Интернет. Подключение может быть двух типов: через телефонную линию (DSL) или оптоволоконный кабель (Fiber optic).

Также доступны такие услуги:

- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).

За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.

## Описание данных

Данные состоят из файлов, полученных из разных источников:

- `personal_new.csv` — персональные данные клиента;

- `contract_new.csv` — информация о договоре;

- `internet_new.csv` — информация об интернет-услугах;

- `phone_new.csv` — информация об услугах телефонии.

Во всех файлах столбец `customerID` содержит код клиента. Информация о договорах актуальна на 1 февраля 2020.

```python
# Функция для формирования информационных отчетов по каждому файлу

def inform_report(list_df):
    mark = ['personal', 'contract', 'internet', 'phone']
    list_reports = []
    list_names = []
    
    for num in range(len(list_df)):
        
        title = 'Information_about_' + mark[num]
        report = sv.analyze([list_df[num], title])
        report_name = 'Common_analysis_' + mark[num] + '.html'
        
        list_reports.append(report)
        list_names.append(report_name)
    
    return dict(zip(list_names, list_reports))
```

```python
# Функция для формирования целевого признака

def create_target(row):
    if row != 'No':
        return 1
    else:
        return 0
```

```python
# Функция для "исправления" признака senior_citizen

def transform_feature(row):
    if row == 0:
        return 'No'
    else:
        return 'Yes'
```

```python
# Функция для подбора наилучших гиперпараметров модели 
# Обучает модель на обучающем наборе данных, возвращает лучшие: модель, параметры и метрику качества (ROC-AUC)

def best_model_params(model, params, features_train, target_train):
    
    gs_model = GridSearchCV(model, param_grid=params, 
                            scoring='roc_auc',
                            cv=10, n_jobs=-1)
    gs_model.fit(features_train, target_train)
    
    best_model = gs_model.best_estimator_
    best_params = gs_model.best_params_
    best_score = gs_model.best_score_
    
    return best_model, best_params, best_score
```

## Исследовательский анализ данных

Создадим новый признак `duration`, в котором отобразим количество дней, в течение которого клиент пользовался услугами:

```python
df['duration'] =  (df['end_date'] - df['begin_date']).dt.days

# проверка
df.columns
```

Такие признаки как `customer_id`, `begin_date` и `end_date` малоинформативны для анализа и дальнейшего обучения модели, удалим их:

```python
df.drop(columns=['customer_id', 'begin_date', 'end_date'], inplace=True)

# проверка
df.columns
```

```python
# Разделение признаков по группам: численные и категориальные 

numerical_features = df.select_dtypes('float').columns.to_list() + ['duration']
categorical_features = df.select_dtypes('object').columns.to_list() # без ц.п.
```

Для численных признаков построим распределения (гистограммы) в разрезе значений целевой переменной (для признаков, характеризующих траты на услуги, строим нормированные гистограммы):

```python
# Визуализация 

fig, ax = plt.subplots(3, 1, figsize=(16, 24)) # задаём сетку графиков

for i, feature in enumerate(numerical_features):
    
    if i < 2:
        plt.subplot(3, 1, i + 1)
        ax = sns.histplot(df[df['exited'] == 0][feature], label='Remained', kde=True, 
                          element='step', color='cadetblue', stat='density', common_norm=False)
        ax = sns.histplot(df[df['exited'] == 1][feature], label='Departed', kde=True,
                          element='step', color='mediumvioletred', stat='density', common_norm=False)
        ax.grid(linewidth=.2, linestyle='--')
        
        # добавляем на графики средние величины платежей (в разрезе ц.п.)
        mean_0 = df[df['exited'] == 0][feature].mean()
        mean_1 = df[df['exited'] == 1][feature].mean()
        plt.axvline(mean_0, color='mediumseagreen', 
                    label='Average spending for remained ({})'.format(np.round(mean_0, 2)),
                    linestyle='--', linewidth=2.5)
        plt.axvline(mean_1, color='hotpink', 
                    label='Average spending for departed ({})'.format(np.round(mean_1, 2)),
                    linestyle='--', linewidth=2.5)
        
        plt.title('Distribution for "{}" '.format(feature), fontsize=14)
        plt.xlabel(feature)
        plt.legend(loc='upper right')
    else:
        plt.subplot(3, 1, i + 1)
        ax = sns.histplot(df[df['exited'] == 0][feature], label='Remained', color='cadetblue', alpha=.45, bins=50)
        ax = sns.histplot(df[df['exited'] == 1][feature], label='Departed', color='mediumvioletred', alpha=.45, bins=50)
        ax.grid(linewidth=.2, linestyle='--')
        plt.title('Distribution for "{}" '.format(feature), fontsize=14)
        plt.legend(loc='upper right')
```

По признаку `monthly_charges` для обоих групп клиентов наблюдается пик в районе значения 20, однако, число оставшихся клиентов в начале пользования услугами (для небольшого месячного платежа) явно преобладает. Для клиентов, которые уходят наблюдается второй пик в районе значений 85-100, то есть можно предположить, что пользователи, имеющие высокий ежемесечный платёж надолго не остаются. Значения в столбце `total_charges` убывают, для ушедших клиентов менее "резко", нежели для оставшихся. Охотнее прекращают быть клиентами те, у кого среднемесячные траты (как и средние общие траты) больше.

Аналогично поступаем и с категориальными признаками (за исключением целевого):

## Подготовка данных

```python
# Удалим признак пола

df.drop(columns='gender', inplace=True)

# проверка
df.columns
```

```python
# Определение признаков, наиболее значимых для модели (вкл. целевой)

# удаляем признак total_charges, на основании чего далее:
    # удаляем признаки коррел. с monthly_charges: streaming_movies, streaming_tv, internet_service, 
                                                # multiple_lines, device_protection
# удаляем признак dependents (коррел. с duration)
    
important_features = ['duration', 'monthly_charges', 'partner', 'senior_citizen', 'type', 'paperless_billing', 
                      'online_security', 'online_backup', 'tech_support', 'exited']
```

```python
# Удаление неинформативных признаков (первый вариант признакового пространства)

df_1 = df[important_features]
df_1.head()
```

```python
# Сохраним признаки и целевой признак (далее ц.п.) в отдельных переменных
# Введём обозначения: X - признаки, y - ц.п. (по аналогии с обозначениями выборок)

X = df_1.drop(columns='exited') # извлекаем признаки

y = df_1['exited'] # извлекаем целевой признак
```

Разобьём наши данные на тренировочную и тестовую выборки:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=STATE, stratify=y)
```

```python
# Проверка

display(X_train.shape, X_test.shape)
display(y_train.shape, y_test.shape)
```

Кодируем категориальные признаки (исп. техники OHE и OE) и масштабируем численные (для линейной модели):

```python
numerical_features = X_train.select_dtypes(include='number').columns.to_list()
numerical_features
```

```python
categorical_features = X_train.select_dtypes(include='object').columns.to_list()
categorical_features
```

```python
# Масштабирование

X_train_scale = X_train.copy()
X_test_scale = X_test.copy()

scaler = StandardScaler()
scaler.fit(X_train_scale[numerical_features])

X_train_scale[numerical_features] = scaler.transform(X_train_scale[numerical_features])
X_test_scale[numerical_features] = scaler.transform(X_test_scale[numerical_features])
```

```python
# Кодирование (для линейной модели)
# Техника OHE

X_train_ohe = X_train_scale.copy()
X_test_ohe = X_test_scale.copy()

encoder_ohe = OneHotEncoder(drop='first', handle_unknown='ignore', sparse=False)
encoder_ohe.fit(X_train_ohe[categorical_features])

X_train_ohe[encoder_ohe.get_feature_names_out()] = encoder_ohe.transform(X_train_ohe[categorical_features])
X_train_ohe = X_train_ohe.drop(categorical_features, axis=1)

X_test_ohe[encoder_ohe.get_feature_names_out()] = encoder_ohe.transform(X_test_ohe[categorical_features])
X_test_ohe = X_test_ohe.drop(categorical_features, axis=1)
```

```python
# Для остальных моделей (кроме линейной)
# Техника OE

X_train_ord = X_train.copy()
X_test_ord = X_test.copy()

encoder_oe = OrdinalEncoder()
encoder_oe.fit(X_train_ord[categorical_features])

X_train_ord[encoder_oe.get_feature_names_out()] = encoder_oe.transform(X_train_ord[categorical_features])
X_test_ord[encoder_oe.get_feature_names_out()] = encoder_oe.transform(X_test_ord[categorical_features])
```
